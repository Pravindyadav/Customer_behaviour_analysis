```python
import pandas as pd
df=pd.read_csv("Customer_shopping_behavior.csv")
df.head()
```

```python
print(df.info())
```

```python
df.describe()            #it gives only summary statistics of numerical value
```

```python
df.describe(include='all')      #it gives summary statistics of both numerical and categorical value
```

```python
#checking the missing values or null values
df.isnull().sum()
```

```python
#cleaning or replacing null values

#df["Review Rating"]=df.groupby('Category')["Review Rating"].transform(lambda x: x.fillna(x.median()))

#or

df["Review Rating"]=df["Review Rating"].fillna(df.groupby('Category')["Review Rating"].transform('median'))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()                  #converting all column into lower case

df.columns=df.columns.str.replace(' ','_')          #it replace space( ) with comma(,)
```

```python
df.info()
df.columns
```

```python
#Renaming specific columns
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
#creating a column age_group
labels=["young","adult","middle_aged","senior"]
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#creating a column purchase_frequency_days

frequency_mapping={'Annually':365,'Weekly':7,'Monthly':30,'Bi-Weekly':14,'Fortnightly':14,'Every 3 Months':90}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```

```python
df[["discount_applied","promo_code_used"]].head(10)
```

```python
#to check both both the columns conatins same values
(df["discount_applied"]==df["promo_code_used"]).all()
```

```python
df=df.drop('promo_code_used', axis=1)
df.columns
```
